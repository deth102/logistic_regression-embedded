```python
import os
import pandas as pd

# ====== CẤU HÌNH ======
INPUT_DIR = r"D:\lt\datn\logistic regression for embedding\Vibration data"
OUTPUT_DIR = r"D:\lt\datn\logistic regression for embedding\Vibration data\Xa_only"

COLUMN_INDEX = 0   # cột đầu tiên (0-based)
# ======================

os.makedirs(OUTPUT_DIR, exist_ok=True)

for filename in os.listdir(INPUT_DIR):
    if not filename.lower().endswith(".csv"):
        continue

    input_path = os.path.join(INPUT_DIR, filename)

    try:
        df = pd.read_csv(input_path)

        # Lấy cột đầu tiên
        first_col = df.iloc[:, COLUMN_INDEX]

        # Tạo tên file mới
        name, ext = os.path.splitext(filename)
        output_filename = f"{name}{ext}"
        output_path = os.path.join(OUTPUT_DIR, output_filename)

        # Lưu ra file mới
        first_col.to_csv(output_path, index=False)

        print(f"[OK] {output_filename}")

    except Exception as e:
        print(f"[ERROR] {filename}: {e}")
```
